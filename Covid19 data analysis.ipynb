```python
# Get a list of unique countries in the DataFrame's index
countries = corona_dataset_aggregated.index

# Create an empty list to store maximum infection rates
max_infection_rates = []

# Iterate over the countries and calculate maximum infection rates
for c in countries:
    # Convert the data to numeric if possible before calculating the difference
    corona_dataset_aggregated.loc[c] = pd.to_numeric(corona_dataset_aggregated.loc[c], errors='coerce')
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())

# Assign the max_infection_rates list as a new column to the DataFrame
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
```
